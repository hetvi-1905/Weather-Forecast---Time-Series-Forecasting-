```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import warnings

warnings.filterwarnings("ignore")

# Load weather dataset (assumed structure as explained above)
data = pd.read_csv('weather_data.csv', parse_dates=['DATE'], index_col='DATE')

# Visualizing the avgtemp column over time
plt.figure(figsize=(10, 5))
plt.plot(data['AvgTemp'], label='Average Temperature')
plt.title('Average Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()

data = data.dropna()

# Split data into training and testing sets (80% train, 20% test)
train_size = int(len(data) * 0.8)
train_data = data[:train_size]
test_data = data[train_size:]

# Fitting ARIMA model on 'avgtemp'
# order=(p,d,q) where p is AR terms, d is differencing, q is MA terms
print("Fitting ARIMA model...")
arima_model = ARIMA(train_data['AvgTemp'], order=(5, 1, 0))  # Adjust order as needed
arima_result = arima_model.fit()

# Forecast using ARIMA model
arima_forecast = arima_result.forecast(steps=len(test_data))

# Fitting SARIMAX model on 'avgtemp'
# order=(p,d,q) and seasonal_order=(P,D,Q,s)
# Seasonal order takes into account seasonality, s is the seasonality period (e.g., 12 for monthly seasonality)
print("Fitting SARIMAX model...")
sarimax_model = SARIMAX(train_data['AvgTemp'], 
                        order=(1, 1, 1), 
                        seasonal_order=(1, 1, 1, 12))
sarimax_result = sarimax_model.fit()

# Forecast using SARIMAX model
sarimax_forecast = sarimax_result.forecast(steps=len(test_data))

# Calculate mean squared error for both models
arima_mse = mean_squared_error(test_data['AvgTemp'], arima_forecast)
sarimax_mse = mean_squared_error(test_data['AvgTemp'], sarimax_forecast)

# Plotting the actual vs forecasted values for both ARIMA and SARIMAX
plt.figure(figsize=(10, 5))
plt.plot(test_data.index, test_data['AvgTemp'], label='Actual Temperature', color='blue')
plt.plot(test_data.index, arima_forecast, label=f'ARIMA Forecast (MSE: {arima_mse:.2f})', color='red', linestyle='--')
plt.plot(test_data.index, sarimax_forecast, label=f'SARIMAX Forecast (MSE: {sarimax_mse:.2f})', color='green', linestyle='--')
plt.title('Temperature Forecasting: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()

# Print MSE for comparison
print(f"ARIMA MSE: {arima_mse:.4f}")
print(f"SARIMAX MSE: {sarimax_mse:.4f}")
```
